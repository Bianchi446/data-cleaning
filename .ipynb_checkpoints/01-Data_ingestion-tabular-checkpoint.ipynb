```python
students = pd.read_csv("data/students-scores.csv")
print(students)
```

```python
def num_score(x):
    to_num = {
        "A+" : 4.3, "A" : 4.2, "A-" : 4.0,
        "B+" : 3.7, "B" : 3.5, "B-" : 3.0,
        "C+" : 2.8, "C" : 2.5, "C-" : 2.1,}
    return x.map(lambda x : to_num.get(x, x))
```

```python
(students
.set_index('Last Name')
.drop('First Name', axis=1)
.apply(num_score)
.T
.plot(title="Student score by year")
.legend(bbox_to_anchor=(1, .75))
)
```

```python
students.melt(
id_vars=["Last Name", "First Name"],
var_name="Level",
value_name="Score"
).set_index(['First Name', 'Last Name', 'Level'])
```

### Using a Regular expression to describe patterns in text 

```python
import re 
pat = re.compile(r'^([^,]+,){5}[^,]*$')
with open("data/big-random.csv") as fh:
    lines = [l.strip().split(',')
            for l in fh if re.match(pat, l)]
pd.DataFrame(lines)
```
